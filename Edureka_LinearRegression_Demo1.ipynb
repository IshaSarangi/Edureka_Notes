##Problem Statement

Given the dataset of advertising, we have to predict the sales based on TV ad expenses

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn.metrics
```

```python
#Load Data
data = pd.read_csv('/content/Advertising.csv')
print(data.head())
```

```python
data.info()
```

```python
#Relationship between TV ads (independent var) and Sales (dependent var)
plt.scatter(data['TV'], data['sales'])
plt.xlabel('TV Ad Expenses')
plt.ylabel('Sales')
plt.show()
```

```python
#Segregate Independent and Dependent variables
x=data['TV'].values.reshape(-1,1)
y=data['sales'].values.reshape(-1,1)
print(type(x))
print(type(y))
```

```python
#print(x)
#print(y)
```

```python
#Train Algorithm
model = LinearRegression()
model.fit(x,y)
```

```python
#Make Prediction
y_pred = model.predict(x)
print(y_pred)
```

```python
plt.scatter(data['TV'],data['sales'])
plt.plot(data['TV'],y_pred,color='red')
plt.xlabel('TV ad exp')
plt.ylabel('Sales')
plt.show()
```

```python
#Coefficients
print(model.coef_) #B1
print(model.intercept_) #B0
```

```python
#Evaluate Model
print("MSE: ", sklearn.metrics.mean_squared_error(y, y_pred))
print("RMSE: ", np.sqrt(sklearn.metrics.mean_squared_error(y, y_pred)))
#print("R2: "), sklearn.metrics.r2_score(y, y_pred)
```

```python
#Prediction on Unknown Data
tv_ad = int(input("Enter TV ad expenses: "))
print(f"Estimated Sales: {model.predict([[tv_ad]])}K")
```
