```python
import pandas as pd
```

```python
#Load Dataset
df=pd.read_csv('/content/nlp_dataset.csv')
df.head()
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#Example1
vect_bow=CountVectorizer(stop_words=stopwords.words('english'),lowercase=True)
text="""Natural language processing enables machines to understand text.\nMachine learning models require large datasets for training."""
text_lines=text.splitlines()
print(text_lines)
bow=vect_bow.fit_transform(text_lines)
pd.DataFrame(bow.toarray(),columns=vect_bow.get_feature_names_out())
```

```python
df_bow=pd.DataFrame(vect_bow.fit_transform(df['Sentence']).toarray(),columns=vect_bow.get_feature_names_out())
print("BOW (Bag of Words) representation:")
df_bow.head()
```

```python
print(vect_bow.get_feature_names_out())
```

```python
#TF-IDF
vect_tfidf=TfidfVectorizer(stop_words=stopwords.words('english'),lowercase=True)
tfidf=vect_tfidf.fit_transform(text_lines)
df_itfidf=pd.DataFrame(tfidf.toarray(),columns=vect_tfidf.get_feature_names_out())
```

```python
print(df_itfidf)
```

```python
#Apply it on corpus
df_tfidf=pd.DataFrame(vect_tfidf.fit_transform(df['Sentence']).toarray(),columns=vect_tfidf.get_feature_names_out())
print("TF-IDF representation:")
df_tfidf.head()
```

```python
from gensim.models import Word2Vec, FastText
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
print(df['Sentence'])
```

```python
sentences = [nltk.word_tokenize(sentences) for sentences in df['Sentence']]
print(sentences)
```

```python
#Train word2vec model
word2vec_model=Word2Vec(sentences,vector_size=50,window=3,sg=0,min_count=1)
#sentences:Training data
#vector_size:each word is represented using this vector size
#window: context windows
#sg=1 using skipgram
#sg=0 using CBOW
```

```python
print("Word2Vec form of the word 'Artificial':")
print(word2vec_model.wv['Artificial'])
```

```python
print("Most similar words (vectors) to 'Artificial':", word2vec_model.wv.most_similar('Artificial'))
```

1. Prepared the sentences --> Tokenize Words
2. Train Word2Vec --> CBOW
3. Extracted Embeddings --> Word Vectors
4. Checked relationship --> similarity and most similar words

##FastText

```python
fasttext_model = FastText(sentences, vector_size=50, window=3, min_count=1)
print("FastText Vector:", fasttext_model.wv['Machine'])
```

```python
print("Similar words to Machine:", fasttext_model.wv.most_similar('Machine'))
```

```python
#OOV Word Test (words not in training data 'Parameter')
print("Word Vector:", fasttext_model.wv['Parameter'])
print("Similar Words:", fasttext_model.wv.most_similar('Parameter'))
```
